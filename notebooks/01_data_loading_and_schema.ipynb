The FD001 training set contains 100 independent engine instances, each representing a complete operational lifecycle under a single operating condition and fault mode.

The absence of missing data reflects the simulated nature of the dataset and differs from real nuclear plant sensor data, where outages, dropouts, and noise are common.

Sensor 1 shows no variance and does not contribute useful information for degradation detection under FD001 conditions.

In this project, “normal operation” is defined as the early cycles of each engine where sensor readings remain within baseline ranges and degradation has not yet caused near-failure behavior. This mirrors nuclear auxiliary cooling pumps, where early-life sensor signals are expected to fluctuate slightly but remain within safe operating thresholds. Later cycles, representing advanced degradation or failure, are reserved for anomaly detection evaluation.

Normal Operation Definition

Normal operation is defined as the early lifecycle behavior of each engine, where sensor readings reflect baseline performance prior to advanced degradation. To ensure a conservative and balanced baseline across engines with varying lifetimes, normal operation is limited to the first 20% of operational cycles, capped at a maximum of 50 cycles per engine. This approach minimizes the risk of including early degradation while maintaining sufficient baseline data for modeling.

```python
#max cycles per engine 
df_max_cycles = df.groupby('engine_id')['cycle'].max()


# applying 20% threshold 
normal_op_cycle_limit = (df_max_cycles * 0.2).clip(upper=50).astype(int)

print(normal_op_cycle_limit)
```

```python
#create boolean column is_normal_operation 

df['is_normal_operation'] = df['cycle'] <= df['engine_id'].map(normal_op_cycle_limit)

print(df['is_normal_operation'])
```

```python
print(df['is_normal_operation'].value_counts())
```

```python
healthy_cycles = df.groupby('engine_id')['is_normal_operation'].sum()

print(healthy_cycles)
```

```python
normal_range = healthy_cycles.max() - healthy_cycles.min()

print(normal_range)
print(healthy_cycles.max())
print(healthy_cycles.min())
```
